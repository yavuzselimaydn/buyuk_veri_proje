```python
import re
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
import re
import nltk
from nltk.corpus import stopwords

# Stopword listesini indir (ilk √ßalƒ±≈ütƒ±rmada gerekli)
nltk.download('stopwords')
stop_words = set(stopwords.words('turkish'))

def clean_text(text):
    # K√º√ß√ºk harfe √ßevir
    text = text.lower()

    # URL temizle
    text = re.sub(r"http\S+|www.\S+", "", text)

    # Mention (@kullanici) temizle
    text = re.sub(r"@\w+", "", text)

    # Hashtag temizle (#hashtag)
    text = re.sub(r"#\w+", "", text)

    # Noktalama kaldƒ±r
    text = re.sub(r"[^\w\s]", "", text)

    # Sayƒ±larƒ± temizle
    text = re.sub(r"\d+", "", text)

    # Emojileri temizle
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # y√ºz ifadeleri
        u"\U0001F300-\U0001F5FF"  # sembol & simgeler
        u"\U0001F680-\U0001F6FF"  # ula≈üƒ±m
        u"\U0001F1E0-\U0001F1FF"  # bayraklar
        "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)

    # Fazla bo≈üluklarƒ± sil
    text = re.sub(r"\s+", " ", text).strip()

    # Tek harfli kelimeleri ve stopword'leri sil
    text = " ".join([
        w for w in text.split()
        if len(w) > 1 and w not in stop_words
    ])

    return text
```

```python

# --- Temizleme SADECE train i√ßin ---
df_train['clean_text'] = df_train['text'].apply(clean_text)
# Bo≈ü metinleri at
df_train = df_train[df_train['clean_text'].str.strip() != '']

print(df_train['clean_text'].apply(lambda x: len(x.split())).describe())

# --- Label Encoding SADECE train'e fit ---
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df_train['label_encoded'] = label_encoder.fit_transform(df_train['label'])

# --- Tokenizer SADECE train'e fit ---
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

MAX_VOCAB = 30000
MAX_LEN = 50

tokenizer = Tokenizer(num_words=MAX_VOCAB, oov_token='<OOV>')
tokenizer.fit_on_texts(df_train['clean_text'])
print("Toplam kelime sayƒ±sƒ±:", len(tokenizer.word_index))

# --- Train verisini sayƒ±salla≈ütƒ±r ---
X_train = tokenizer.texts_to_sequences(df_train['clean_text'])
X_train = pad_sequences(X_train, maxlen=MAX_LEN, padding='post')
y_train = df_train['label_encoded'].values

# --- Test verisini bu tokenizer ve encoder ile D√ñN√ú≈ûT√úR (fit yok) ---
df_test['clean_text'] = df_test['text'].apply(clean_text)
df_test['label_encoded'] = label_encoder.transform(df_test['label'])

X_test = tokenizer.texts_to_sequences(df_test['clean_text'])
X_test = pad_sequences(X_test, maxlen=MAX_LEN, padding='post')
y_test = df_test['label_encoded'].values
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def evaluate_model(y_true, y_pred, label_encoder, title="Model"):
    """
    y_true: ger√ßek etiketler (sayƒ±sal)
    y_pred: tahmin edilen etiketler (sayƒ±sal)
    label_encoder: LabelEncoder nesnesi
    title: grafikte g√∂sterilecek model adƒ±
    """
    class_names = label_encoder.classes_

    # 1. Performans metrikleri
    report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
    df_report = pd.DataFrame(report).transpose()

    # 2. Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title(f"{title} - Confusion Matrix")
    plt.xlabel("Tahmin")
    plt.ylabel("Ger√ßek")
    plt.tight_layout()
    plt.show()

    # 3. Precision, Recall, F1-score Barplot
    df_plot = df_report.loc[class_names, ['precision', 'recall', 'f1-score']]
    df_plot.plot(kind='bar', figsize=(8, 5), colormap='Set2')
    plt.title(f"{title} - Sƒ±nƒ±f Bazlƒ± Metrikler")
    plt.ylabel("Skor")
    plt.ylim(0, 1.05)
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()

    # 4. Genel metrikler
    acc = accuracy_score(y_true, y_pred)
    macro_f1 = df_report.loc["macro avg", "f1-score"]
    weighted_f1 = df_report.loc["weighted avg", "f1-score"]

    print(f"üìä {title} Genel Sonu√ßlar:")
    print(f"Accuracy          : {acc:.4f}")
    print(f"Macro F1-Score    : {macro_f1:.4f}")
    print(f"Weighted F1-Score : {weighted_f1:.4f}")
```

**LSTM + Attention**

```python
import tensorflow as tf
from tensorflow.keras.layers import Layer, Embedding, LSTM, Dense, Dropout
from tensorflow.keras.models import Model
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

class_weights = compute_class_weight(class_weight="balanced", classes=np.unique(y_train), y=y_train)
class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}
print("Sƒ±nƒ±f Aƒüƒ±rlƒ±klarƒ±:", class_weight_dict)

class Attention(tf.keras.layers.Layer):
    def __init__(self):
        super(Attention, self).__init__()
        self.dense = tf.keras.layers.Dense(1)

    def call(self, inputs):
        scores = tf.nn.softmax(self.dense(inputs), axis=1)
        context = tf.reduce_sum(scores * inputs, axis=1)
        return context

MAX_VOCAB = 30000  # Kendi deƒüerinle deƒüi≈ütir
MAX_LEN = 50

input_layer = tf.keras.Input(shape=(MAX_LEN,))
x = Embedding(input_dim=MAX_VOCAB, output_dim=128)(input_layer)
x = LSTM(64, return_sequences=True)(x)
x = Attention()(x)
x = Dropout(0.5)(x)
x = Dense(64, activation='relu')(x)
output_layer = Dense(3, activation='softmax')(x)

model_attn = Model(inputs=input_layer, outputs=output_layer)

model_attn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

from tensorflow.keras.callbacks import EarlyStopping

# Early stopping callback tanƒ±mƒ±
early_stop = EarlyStopping(
    monitor='val_loss',     # Doƒürulama kaybƒ±nƒ± izle
    patience=2,             # √úst √ºste 2 epoch boyunca iyile≈üme olmazsa durdur
    restore_best_weights=True # En iyi model aƒüƒ±rlƒ±klarƒ±nƒ± geri y√ºkle
)

# model eƒüitimi
history_attn = model_attn.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=20,
    batch_size=512,
    class_weight=class_weight_dict,
    callbacks=[early_stop],   # CALLBACK EKLENDƒ∞
    verbose=1
)

y_pred_probs = model_attn.predict(X_test)
y_pred_attn = np.argmax(y_pred_probs, axis=1)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_attn, target_names=label_encoder.classes_))
```

```python
plt.figure(figsize=(10,6))
bars = plt.bar(model_names, precision, color='mediumseagreen')
plt.title("Model Precision (Macro) Kar≈üƒ±la≈ütƒ±rmasƒ±")
plt.ylabel("Precision")
plt.ylim(0.75, 1.0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
             f'{bar.get_height():.2f}', ha='center')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
